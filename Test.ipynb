```python
import numpy as np
import json
import requests
def riskAnalysis(company_name):
    company_code = code_convertor(company_name)
    r = requests.post('http://www.blackrock.com/tools/hackathon/performance?identifiers=' + company_code, verify=False)
    print(r.json())
def code_convertor(company_name):
    company_dict = {
        'google':'GOOGL',
        'apple':'AAPL',
        'adobe':'ADBE',
        'applied materials':'AMAT',
        'amazon':'AMZN',
        'broadcom':'AVGO',
        'cisco':'CSCO',
        'citrix':'CTXS',
        'comcast':'CMCSA',
        'ebay':'EBAY',
        'expedia':'EXPE',
        'facebook':'FB',
        'intel':'INTC',
        'microsoft':'MSFT',
        'netflix':'NFLX',
        'nvidia':'NVDA',
        'paypal':'PYPL',
        'qualcomm':'QCOM',
        'starbucks':'SBUX',
        'tesla':'TSLA'}
    for key in company_dict:
        if company_name == key:
            return company_dict[key]
```
