```python
import numpy as np
import json
import requests
def riskAnalysis(company_name_list):
    inter_company_code = ''
    for i in range(0, len(company_name_list)):
        company_code = code_convertor(company_name_list[i])
        if(company_code is not None):
            if i == 0:
                inter_company_code += company_code
            else:
                inter_company_code += '%2C' + company_code
                
        
    r = requests.post('http://www.blackrock.com/tools/hackathon/performance?identifiers=' + inter_company_code, verify=False)
    print(r.json())
def code_convertor(company_name):
    company_dict = {
        'google':'GOOGL',
        'apple':'AAPL',
        'adobe':'ADBE',
        'applied materials':'AMAT',
        'amazon':'AMZN',
        'broadcom':'AVGO',
        'cisco':'CSCO',
        'citrix':'CTXS',
        'comcast':'CMCSA',
        'ebay':'EBAY',
        'expedia':'EXPE',
        'facebook':'FB',
        'intel':'INTC',
        'microsoft':'MSFT',
        'netflix':'NFLX',
        'nvidia':'NVDA',
        'paypal':'PYPL',
        'qualcomm':'QCOM',
        'starbucks':'SBUX',
        'tesla':'TSLA'}
    for key in company_dict:
        if company_name == key:
            return company_dict[key]
```
